```python
df=data
y= df['Ep'].copy()
x = df.drop('Ep', axis=1)
x=x.drop('No.', axis=1)
test_y=test['Ep'].copy()
test_x = test.drop('Ep', axis=1)
test_x = test_x.drop('No.', axis=1)
type(y)
x
test_x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import plot_importance
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error


# 参数
model = xgb.XGBRegressor(max_depth=10,
                        learning_rate=0.1,
                        n_estimators=300,
                        objective='reg:squarederror', 
                        booster='gbtree',
                        gamma=0.1,
                        min_child_weight=1,
                        subsample=0.8,
                        colsample_bytree=0.9,
                        reg_alpha=0,
                        reg_lambda=1,
                        )
MAE=[]
RMSE=[]
R2=[]
for i in range(10):
    # train an XGBoost model
    X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2)
    model.fit(X_train,Y_train)
    # 
    ans = model.predict(X_test)
    mse = mean_squared_error(Y_test,ans)**0.5
    r2=r2_score(Y_test,ans)
    mae=mean_absolute_error(Y_test,ans)
    R2.append(r2)
    RMSE.append(mse)
    MAE.append(mae)

print('MAE:', MAE)
print('RMSE', RMSE)
print('R2:',R2)
# mean
MAE_mean = np.mean(MAE)
RMSE_mean = np.mean(RMSE)
R2_mean = np.mean(R2)
# std
MAE_std = np.std(MAE,ddof=1)
RMSE_std = np.std(RMSE,ddof=1)
R2_std = np.std(R2,ddof=1)
print('mean:', MAE_mean,RMSE_mean,R2_mean)
print('std', MAE_std,RMSE_std,R2_std)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
# AdaBoost
AdaBoost_MAE=[]
AdaBoost_RMSE=[]
AdaBoost_R2=[]
for i in range(10):
    # train an XGBoost model
    X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2)
    model = AdaBoostRegressor(DecisionTreeRegressor(max_depth=15),n_estimators=500, learning_rate=0.2, loss='linear', random_state=None)
    model.fit(X_train,Y_train)
    # 
    ans = model.predict(X_test)
    mse = mean_squared_error(Y_test,ans)**0.5
    r2=r2_score(Y_test,ans)
    mae=mean_absolute_error(Y_test,ans)
    AdaBoost_R2.append(r2)
    AdaBoost_RMSE.append(mse)
    AdaBoost_MAE.append(mae)

print('AdaBoost_MAE:', AdaBoost_MAE)
print('AdaBoost_RMSE', AdaBoost_RMSE)
print('AdaBoost_R2:',AdaBoost_R2)
# mean
AdaBoost_MAE_mean = np.mean(AdaBoost_MAE)
AdaBoost_RMSE_mean = np.mean(AdaBoost_RMSE)
AdaBoost_R2_mean = np.mean(AdaBoost_R2)
# std
AdaBoost_MAE_std = np.std(AdaBoost_MAE,ddof=1)
AdaBoost_RMSE_std = np.std(AdaBoost_RMSE,ddof=1)
AdaBoost_R2_std = np.std(AdaBoost_R2,ddof=1)
print('mean:', AdaBoost_MAE_mean,AdaBoost_RMSE_mean,AdaBoost_R2_mean)
print('std', AdaBoost_MAE_std,AdaBoost_RMSE_std,AdaBoost_R2_std)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
# AdaBoost
RF_MAE=[]
RF_RMSE=[]
RF_R2=[]
for i in range(10):
    # train an XGBoost model
    X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2)
    model = RandomForestRegressor(n_estimators=200, criterion='squared_error',max_depth=15, random_state=None)
    model.fit(X_train,Y_train)

    ans = model.predict(X_test)
    mse = mean_squared_error(Y_test,ans)**0.5
    r2=r2_score(Y_test,ans)
    mae=mean_absolute_error(Y_test,ans)
    RF_R2.append(r2)
    RF_RMSE.append(mse)
    RF_MAE.append(mae)

print('RF_MAE:', RF_MAE)
print('RF_RMSE', RF_RMSE)
print('RF_R2:',RF_R2)
# mean
RF_MAE_mean = np.mean(RF_MAE)
RF_RMSE_mean = np.mean(RF_RMSE)
RF_R2_mean = np.mean(RF_R2)
# std
RF_MAE_std = np.std(RF_MAE,ddof=1)
RF_RMSE_std = np.std(RF_RMSE,ddof=1)
RF_R2_std = np.std(RF_R2,ddof=1)
print('mean:', RF_MAE_mean,RF_RMSE_mean,RF_R2_mean)
print('std', RF_MAE_std,RF_RMSE_std,RF_R2_std)
```
